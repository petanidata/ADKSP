## 🧹 3. Text Preprocessing

```python
# 3.1 Stopwords Bahasa Indonesia
STOPWORDS_ID = set([
    'yang', 'dan', 'di', 'ke', 'dari', 'dengan', 'untuk', 'pada', 'adalah', 'ini',
    'itu', 'atau', 'juga', 'oleh', 'akan', 'sudah', 'telah', 'dalam', 'tidak',
    'ada', 'sehingga', 'namun', 'berkat', 'akibat', 'melalui', 'seiring', 'atas',
    'besar', 'sangat', 'cukup', 'baik', 'jauh', 'masih', 'saja', 'hanya', 'sama',
    'lebih', 'bagi', 'setelah', 'saat', 'agar', 'menjadi', 'kami', 'kita'
])

def preprocess_teks(teks):
    """Preprocessing teks Bahasa Indonesia."""
    # 1. Lowercase
    teks = teks.lower()
    # 2. Hapus tanda baca dan angka
    teks = re.sub(r'[^a-zA-Z\s]', ' ', teks)
    # 3. Tokenisasi (split)
    tokens = teks.split()
    # 4. Hapus stopwords
    tokens = [t for t in tokens if t not in STOPWORDS_ID and len(t) > 2]
    # 5. Stemming (jika Sastrawi tersedia)
    if SASTRAWI:
        tokens = [stemmer.stem(t) for t in tokens]
    return tokens

# Terapkan preprocessing
df_teks['Tokens'] = df_teks['Narasi'].apply(preprocess_teks)
df_teks['Teks_Bersih'] = df_teks['Tokens'].apply(lambda x: ' '.join(x))

# Contoh hasil
print('CONTOH PREPROCESSING')
print('=' * 60)
print('Teks Asli:')
print(df_teks['Narasi'].iloc[0][:120] + '...')
print()
print('Setelah Preprocessing:')
print(df_teks['Tokens'].iloc[0])
```

```python
# 3.2 Analisis Frekuensi Kata
all_tokens = [token for tokens in df_teks['Tokens'] for token in tokens]
word_freq = Counter(all_tokens)
top_words = word_freq.most_common(20)

words, freqs = zip(*top_words)

fig, ax = plt.subplots(figsize=(11, 5))
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(words)))
bars = ax.barh(list(reversed(words)), list(reversed(freqs)), color=list(reversed(colors)))
ax.set_xlabel('Frekuensi', fontsize=11)
ax.set_title('Top 20 Kata Paling Sering Muncul\ndalam Narasi Keuangan PEMDA', fontweight='bold', fontsize=13)
for bar, val in zip(bars, reversed(freqs)):
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, str(val), va='center')
plt.tight_layout()
plt.show()
```

## 📊 5. TF-IDF

```python
# Hitung TF-IDF
vectorizer = TfidfVectorizer(max_features=30, min_df=1, max_df=0.9)
tfidf_matrix = vectorizer.fit_transform(df_teks['Teks_Bersih'])
feature_names = vectorizer.get_feature_names_out()

df_tfidf = pd.DataFrame(tfidf_matrix.toarray(),
                         index=df_teks['Kode_Pemda'],
                         columns=feature_names)

print('Matrix TF-IDF (baris=dokumen, kolom=term):')
print(f'Dimensi: {df_tfidf.shape}')
df_tfidf.round(3)
```

```python
# Heatmap TF-IDF
# Pilih top kata (dengan skor TF-IDF total tertinggi)
top_terms_idx = df_tfidf.sum(axis=0).nlargest(20).index
df_tfidf_top = df_tfidf[top_terms_idx]

fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(df_tfidf_top, annot=True, fmt='.2f', cmap='YlOrRd',
            ax=ax, linewidths=0.3, cbar_kws={'label': 'TF-IDF Score'})
ax.set_title('Heatmap TF-IDF: Bobot Kata per Dokumen\n(semakin tinggi = kata semakin unik/penting di dokumen tersebut)',
             fontweight='bold')
ax.set_xlabel('Kata (Term)')
ax.set_ylabel('Kode PEMDA')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## 😊 6. Analisis Sentimen Sederhana

```python
# Analisis sentimen berbasis kamus (Lexicon-based)
KATA_POSITIF = [
    'berhasil', 'meningkat', 'surplus', 'efisiensi', 'optimal', 'baik',
    'melebihi', 'kompetitif', 'pemberdayaan', 'peningkatan', 'signifikan',
    'wajar', 'penuh', 'mencapai', 'investasi', 'reformasi'
]

KATA_NEGATIF = [
    'temuan', 'kerugian', 'defisit', 'penurunan', 'keterlambatan', 'kelemahan',
    'fiktif', 'rasionalisasi', 'rendah', 'minim', 'hanya', 'terpaksa',
    'terlambat', 'masalah', 'gagal', 'kekurangan'
]

def analisis_sentimen(teks):
    teks_lower = teks.lower()
    skor_pos = sum(1 for k in KATA_POSITIF if k in teks_lower)
    skor_neg = sum(1 for k in KATA_NEGATIF if k in teks_lower)
    skor_total = skor_pos - skor_neg

    if skor_total > 1:
        sentimen = 'Positif'
    elif skor_total < -1:
        sentimen = 'Negatif'
    else:
        sentimen = 'Netral'

    return pd.Series({'Skor_Positif': skor_pos, 'Skor_Negatif': skor_neg,
                      'Skor_Total': skor_total, 'Sentimen': sentimen})

sentimen_df = df_teks['Narasi'].apply(analisis_sentimen)
df_teks = pd.concat([df_teks, sentimen_df], axis=1)

print('HASIL ANALISIS SENTIMEN')
print('=' * 50)
print(df_teks[['Kode_Pemda', 'Skor_Positif', 'Skor_Negatif', 'Skor_Total', 'Sentimen']].to_string(index=False))
```

```python
# Visualisasi sentimen
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart skor sentimen
colors = ['#27ae60' if s == 'Positif' else '#e74c3c' if s == 'Negatif' else '#95a5a6'
          for s in df_teks['Sentimen']]
bars = axes[0].bar(df_teks['Kode_Pemda'], df_teks['Skor_Total'], color=colors, edgecolor='white')
axes[0].axhline(0, color='black', linewidth=0.8)
axes[0].set_xlabel('Kode PEMDA')
axes[0].set_ylabel('Skor Sentimen')
axes[0].set_title('Skor Sentimen per PEMDA\n(hijau=positif, merah=negatif)', fontweight='bold')
axes[0].tick_params(axis='x', rotation=30)

# Distribusi sentimen
sent_counts = df_teks['Sentimen'].value_counts()
colors_pie = ['#27ae60', '#95a5a6', '#e74c3c']
axes[1].pie(sent_counts.values, labels=sent_counts.index, colors=colors_pie[:len(sent_counts)],
            autopct='%1.0f%%', startangle=90)
axes[1].set_title('Distribusi Sentimen\nNarasi PEMDA', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Document Similarity — Dokumen mana yang paling mirip?
similarity_matrix = cosine_similarity(tfidf_matrix)
df_sim = pd.DataFrame(similarity_matrix,
                       index=df_teks['Kode_Pemda'],
                       columns=df_teks['Kode_Pemda'])

# Mask diagonal
np.fill_diagonal(similarity_matrix, 0)

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(df_sim, annot=True, fmt='.2f', cmap='Blues', ax=ax,
            linewidths=0.3, cbar_kws={'label': 'Cosine Similarity'})
ax.set_title('Document Similarity\n(berdasarkan TF-IDF — nilai > 0.3 = sangat mirip)', fontweight='bold')
plt.tight_layout()
plt.show()

# Temukan pasangan paling mirip
max_idx = np.unravel_index(similarity_matrix.argmax(), similarity_matrix.shape)
print(f'\nDokumen paling mirip:')
print(f'  {df_teks["Kode_Pemda"].iloc[max_idx[0]]} ↔ {df_teks["Kode_Pemda"].iloc[max_idx[1]]}')
print(f'  Similarity Score: {similarity_matrix[max_idx]:.4f}')
```
